```python
import pandas as pd

import mysql.connector
from sqlalchemy import create_engine
from getpass import getpass
import pymysql
```

```python
password = getpass.getpass()
```

```python
db_user = 'root'
db_password = password
db_host = '127.0.0.1'
db_name = 'sakila'

engine = create_engine(f'mysql+pymysql://{db_user}:{db_password}@{db_host}/{db_name}')
```

```python
def rentals_month(engine, month, year):

    query = f"""
    SELECT r.rental_id, r.rental_date, r.inventory_id, r.customer_id, f.title AS film_title
    FROM rental r
    JOIN inventory i ON r.inventory_id = i.inventory_id
    JOIN film f ON i.film_id = f.film_id
    WHERE MONTH(r.rental_date) = {month} AND YEAR(r.rental_date) = {year};
    """
    
    df = pd.read_sql(query, engine)
    return df
```

```python
month = 6
year = 2005

rental_data = rentals_month(engine, month, year)


rental_data.head(50)
```

```python

def rental_count_month(rental_df, month, year):
    # Group the data by customer_id and count the number of rentals for each customer
    rental_counts = rental_df.groupby('customer_id').size().reset_index(name=f'rentals_{str(month).zfill(2)}_{year}')
    
    return rental_counts
```

```python
rental_counts_june_2005 = rental_count_month(rental_data, month, year)


rental_counts_june_2005.head(50)
```

```python
month = 7
year = 2005
```

```python
rental_data = rentals_month(engine, month, year)
```

```python
rental_counts_july_2005 = rental_count_month(rental_data, month, year)
rental_counts_july_2005.head(50)
```

```python
import pandas as pd

def compare_rentals(df1, df2, month1, year1, month2, year2):
    # Merge the two DataFrames on 'customer_id'
    df_combined = pd.merge(df1, df2, on='customer_id', how='outer', suffixes=(f'_{str(month1).zfill(2)}_{year1}', f'_{str(month2).zfill(2)}_{year2}'))
    
    # Fill missing values with 0
    df_combined.fillna(0, inplace=True)
    
    # Dynamically get the rental column names
    col1 = f'rentals_{str(month1).zfill(2)}_{year1}'
    col2 = f'rentals_{str(month2).zfill(2)}_{year2}'
    
    # Convert rental count columns to integers
    df_combined[col1] = df_combined[col1].astype(int)
    df_combined[col2] = df_combined[col2].astype(int)
    
    df_combined['difference'] = df_combined[col2] - df_combined[col1]
    
    return df_combined
```

```python

rental_data_june = rentals_month(engine, 6, 2005)
rental_counts_june = rental_count_month(rental_data_june, 6, 2005)


rental_data_july = rentals_month(engine, 7, 2005)
rental_counts_july = rental_count_month(rental_data_july, 7, 2005)


comparison_df = compare_rentals(rental_counts_june, rental_counts_july, 6, 2005, 7, 2005)


display(comparison_df.head())
```
